# Title Simplifier and Seniority Functions

1. **`title_simplifier` function**: Simplifies job titles by categorizing them into common roles like Data Scientist, Data Engineer, Analyst, etc.
2. **`seniority` function**: Identifies the seniority level based on the title, categorizing it as Senior, Junior, or not applicable.

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'machine learning'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr.' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

# Applying Functions to the DataFrame

We will apply the `title_simplifier` and `seniority` functions to the DataFrame to categorize job titles and seniority levels.

```python
df['job_simp'] = df['Job Title'].apply(title_simplifier)
```

```python
df.job_simp.value_counts()
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```

# Converting Hourly Wage to Annual Salary

We will convert the hourly wage to an annual salary by assuming a standard full-time work schedule (40 hours/week, 52 weeks/year).

```python
df['min_salary'] = df.apply(
    lambda x: x['min_salary']*2 if pd.notna(x['min_salary']) and x['hourly'] == 1 else x['min_salary'],
    axis=1
)
df['max_salary'] = df.apply(
    lambda x: x['max_salary']*2 if pd.notna(x['max_salary']) and x['hourly'] == 1 else x['max_salary'],
    axis=1
)
```
